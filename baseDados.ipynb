```python
import pandas as pd 

df = pd.read_csv('custo_de_vida_nos_paises_2024.csv')


# Renomear as colunas para português
df.columns = [
    'Rank', 
    'País', 
    'Índice de Custo de Vida', 
    'Índice de Aluguel', 
    'Índice de Custo de Vida + Aluguel', 
    'Índice de Supermercados', 
    'Índice de Restaurantes', 
    'Índice de Poder de Compra Local'
]

# Exibir o cabeçalho do DataFrame atualizadoz

df.head()

```

```python
top_5_countries = df.nlargest(5, 'Índice de Custo de Vida')
bottom_5_countries = df.nsmallest(5, 'Índice de Custo de Vida')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotar o gráfico de barras com os títulos em português
plt.figure(figsize=(20, 6))
df.set_index('País')['Índice de Custo de Vida'].plot(kind='bar')
plt.title('Índice de Custo de Vida por País')
plt.xlabel('País')
plt.ylabel('Índice de Custo de Vida')
plt.show()
```

```python
# Plotar o gráfico de dispersão com os títulos em português
plt.figure(figsize=(20, 6))
plt.scatter(df['Índice de Custo de Vida'], df['Índice de Aluguel'])
plt.title('Índice de Custo de Vida vs Índice de Aluguel')
plt.xlabel('Índice de Custo de Vida')
plt.ylabel('Índice de Aluguel')
plt.show()
```

```python
# Calcular a correlação entre os índices
correlation_matrix = df[['Índice de Custo de Vida', 'Índice de Aluguel', 'Índice de Supermercados', 'Índice de Restaurantes', 'Índice de Poder de Compra Local']].corr()

# Plotar o heatmap de correlação com o título em português
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlação dos Índices de Custo de Vida')
plt.show()
```

## Índice de Custo de Vida versus Poder de Compra Local

```python
# Configurar o estilo do seaborn
sns.set(style="whitegrid")

# Scatter plot do Índice de Custo de Vida vs Índice de Poder de Compra Local
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Índice de Custo de Vida', y='Índice de Poder de Compra Local', data=df, hue='País', palette='viridis', s=100, alpha=0.7, edgecolor='k')

# Adicionar títulos e rótulos
plt.title('Relação entre Índice de Custo de Vida e Índice de Poder de Compra Local', fontsize=16)
plt.xlabel('Índice de Custo de Vida', fontsize=14)
plt.ylabel('Índice de Poder de Compra Local', fontsize=14)

# Colocar a legenda abaixo do gráfico em linha horizontal
plt.legend(bbox_to_anchor=(0.5, -0.2), loc='upper center', ncol=4, borderaxespad=0.)

# Mostrar o gráfico
plt.show()
```

## TOP 5

```python
# Identificar os top 5 países pelo Índice de Custo de Vida
top_5_countries = df.nlargest(5, 'Índice de Custo de Vida')
bottom_5_countries = df.nsmallest(5, 'Índice de Custo de Vida')

# Visualizar os top 5 países pelo Índice de Custo de Vida
plt.figure(figsize=(10, 6))
sns.barplot(x='Índice de Custo de Vida', y='País', data=top_5_countries, palette='viridis')

# Adicionar títulos e rótulos
plt.title('Top 5 Países por Índice de Custo de Vida', fontsize=16)
plt.xlabel('Índice de Custo de Vida', fontsize=14)
plt.ylabel('País', fontsize=14)

# Mostrar o gráfico
plt.show()

# Visualizar os bottom 5 países pelo Índice de Custo de Vida
plt.figure(figsize=(10, 6))
sns.barplot(x='Índice de Custo de Vida', y='País', data=bottom_5_countries, palette='viridis')

# Adicionar títulos e rótulos
plt.title('Bottom 5 Países por Índice de Custo de Vida', fontsize=16)
plt.xlabel('Índice de Custo de Vida', fontsize=14)
plt.ylabel('País', fontsize=14)

# Mostrar o gráfico
plt.show()
```
